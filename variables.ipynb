```python
df = pd.read_csv('prueba.csv')
```

```python
# Ejemplito

consulta = df[df['Alcaldia'] == 'San Pedro Tlaquepaque'][['CP', 'P_NIVSOCEC' ]].reset_index()
consulta["Calificacion"] = ""
qs = consulta['P_NIVSOCEC'].quantile([.2, .4, .6, .8]).values

i=0
while i < len(consulta):
        if consulta['P_NIVSOCEC'][i] <= qs[0]:
            consulta.loc[i, "Calificacion"] = "6"
        elif (consulta['P_NIVSOCEC'][i] <=qs[1]) & (consulta['P_NIVSOCEC'][i] > qs[0]):
            consulta.loc[i, "Calificacion"] = "7"
        elif (consulta['P_NIVSOCEC'][i] <=qs[2]) & (consulta['P_NIVSOCEC'][i] > qs[1]):
            consulta.loc[i, "Calificacion"] = "8"
        elif (consulta['P_NIVSOCEC'][i] <=qs[3]) & (consulta['P_NIVSOCEC'][i] > qs[2]):
            consulta.loc[i, "Calificacion"] = "9"
        else: 
            consulta.loc[i, "Calificacion"] = "10"
         
        i += 1

consulta.head()     
```
