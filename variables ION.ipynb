```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Variables_Municipios.csv')
df
```

```python
len(df['CP'].unique())
```

```python
variables = ['P_NIVSOCEC', 'Sum_Servicios', 'Num_Celulares','Num_Autos', 'Empresas', 'Empleos']
alcaldia = df.Alcaldia.unique()
quantiles = []

for w in variables:
    nombre = "Cal_"+str(w)
    df[nombre] = ""


for x in alcaldia:   
    for y in variables:
        
        consulta = df[df['Alcaldia'] == str(x)][[str(y)]].reset_index()
        qs = consulta[y].quantile([.2, .4, .6, .8, 1]).values
        quantiles.append(qs)
        nombre = "Cal_"+str(y)
    
        for a,b in consulta.itertuples(index=False):
            if b <= qs[0]:
                df.loc[a, str(nombre)] = "1"
            elif (b <= qs[1]) & (b > qs[0]):
                df.loc[a, str(nombre)] = "2"
            elif (b <= qs[2]) & (b > qs[1]):
                df.loc[a, str(nombre)] = "3"
            elif (b <= qs[3]) & (b > qs[2]):
                df.loc[a, str(nombre)] = "4"
            elif b <= qs[4]:
                df.loc[a, str(nombre)] = "5"
            else:
                df.loc[a, str(nombre)] = "0"
```
